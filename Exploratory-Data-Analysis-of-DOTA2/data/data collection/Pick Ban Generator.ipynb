```python
matchid = pd.read_csv('international_2016.csv')
matchid.head()
df1 = pd.DataFrame()
data = {}
for m in matchid['0']:
    z ='https://api.steampowered.com/IDOTA2Match_570/GetMatchDetails/V001/?match_id='+str(m)+'&key=22B7FC208EA5722620B22DCEA99C638A'
    source = requests.get(z)
    matchdetails = json.loads(source.text)['result']
    ##print(matchdetails)
    data = {}
    data['match_id'] = matchdetails['match_id']
    ban1 = ban2 = pick1 = pick2 = 0
    for a in matchdetails['picks_bans']:
        if(a['is_pick'] == False):
            if(a['team'] == 0):
                ban1 +=1
                data['Ban_T1_'+str(ban1)] = a['hero_id']
            else:
                ban2 +=1
                data['Ban_T2_'+str(ban2)] = a['hero_id']
                
        else:
            if(a['team'] == 0):
                pick1 +=1
                data['Pick_T1_'+str(pick1)] = a['hero_id']
            else:
                pick2 +=1
                data['Pick_T2_'+str(pick2)] = a['hero_id']
    ##print(matchdetails['radiant_win'])
    data['result'] = 'Radiant' if matchdetails['radiant_win'] == True else 'Dire'
    ##print(data)
    df1 = df1.append(data,ignore_index =True)
    
    

for a in list(df1.columns):
    print(a,list(df1.columns).index(a))

colorder = [20,21] + [x for x in range(20)]
df1 = df1.iloc[:,colorder]
df1.to_csv('International_2016_drafting.csv',index= False)
```
