```python
temp=[]
for i in range(len(df)):
    if(df.loc[i,'Result']=='Radiant'):
        for j in range(1,6):
            temp.append(df.loc[i,'Player_'+str(j)+'_hero_id'])
    else:
        for j in range(6,11):
            temp.append(df.loc[i,'Player_'+str(j)+'_hero_id'])

temp1 ,temp2 = np.unique(np.array(temp),return_counts=True)
hero_win_count_2016 = {temp1[i]:temp2[i] for i in range(len(temp1))}



hero_win_count_2016 = {x:y for x,y in sorted(hero_win_count_2016.items(), key = lambda item:item[0])}
```
