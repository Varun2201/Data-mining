```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
#Sets the plots to be displayed and saved in notebook
%matplotlib inline 
plt.rcParams["figure.figsize"] = (20,10)
```

```python
df  = pd.read_csv('../data/match-details/2016.csv')
df1 = pd.read_csv('../data/match-details/2017.csv')
df2 = pd.read_csv('../data/match-details/2018.csv')
df3 = pd.read_csv('../data/match-details/2019.csv')
df4 = pd.read_csv('../data/match-details/2021.csv')
```

```python
KPM_2016 = ((df['Radiant_score']+df['Dire_score'])/df['Duration']).to_numpy()
KPM_2017 = ((df1['Radiant_score']+df1['Dire_score'])/df1['Duration']).to_numpy()
KPM_2018 = ((df2['Radiant_score']+df2['Dire_score'])/df2['Duration']).to_numpy()
KPM_2019 = ((df3['Radiant_score']+df3['Dire_score'])/df3['Duration']).to_numpy()
KPM_2021 = ((df4['Radiant_score']+df4['Dire_score'])/df4['Duration']).to_numpy()
```

```python
fig , ax = plt.subplots(1)



ax.plot(['2016','2017','2018','2019','2021'],[np.average(KPM_2016*60),np.average(KPM_2017*60),np.average(KPM_2018*60),np.average(KPM_2019*60),np.average(KPM_2021*60)],color='Magenta')

```

```python
fig , ax = plt.subplots(1)

ax.plot(['2016','2017','2018','2019','2021'],[np.average(df['Duration']/60),np.average(df1['Duration']/60),np.average(df2['Duration']/60),np.average(df3['Duration']/60),np.average(df4['Duration']/60)],color='orange')
print([np.average(df['Duration']/60),np.average(df1['Duration']/60),np.average(df2['Duration']/60),np.average(df3['Duration']/60),np.average(df4['Duration']/60)])
```
