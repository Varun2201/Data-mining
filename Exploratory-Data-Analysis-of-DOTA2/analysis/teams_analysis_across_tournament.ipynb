```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
import warnings
warnings.simplefilter('ignore')

%matplotlib inline 
plt.rcParams["figure.figsize"] = (30,30)
```

```python
df  = pd.read_csv('../output/Team-analysis2016.csv')
df1 = pd.read_csv('../output/Team-analysis2017.csv')
df2 = pd.read_csv('../output/Team-analysis2018.csv')
df3 = pd.read_csv('../output/Team-analysis2019.csv')
df4 = pd.read_csv('../output/Team-analysis2021.csv')
```

```python
df.sort_values('Team_name',inplace=True)
df1.sort_values('Team_name',inplace=True)
df2.sort_values('Team_name',inplace=True)
df3.sort_values('Team_name',inplace=True)
df4.sort_values('Team_name',inplace=True)
```

```python
plt.subplot(5,1,1)
plt.barh(df['Team_name'],df['Total_kills'],color='Red')
plt.title('Total_kills')
plt.subplot(5,1,2)
plt.barh(df1['Team_name'],df1['Total_kills'],color='Blue')
plt.title('Total_kills')
plt.subplot(5,1,3)
plt.barh(df2['Team_name'],df2['Total_kills'],color='Black')
plt.title('Total_kills')
plt.subplot(5,1,4)
plt.barh(df3['Team_name'],df3['Total_kills'],color='Green')
plt.title('Total_kills')
plt.subplot(5,1,5)
plt.barh(df4['Team_name'],df4['Total_kills'],color='Purple')
plt.title('Total_kills')
print()
```

```python
plt.subplot(5,1,1)
plt.title('Unique heros played')
plt.barh(df['Team_name'],df['Unique_heros_played'],color='Red')
plt.subplot(5,1,2)
plt.title('Unique heros played')
plt.barh(df1['Team_name'],df1['Unique_heros_played'],color='Blue')

plt.subplot(5,1,3)
plt.title('Unique heros played')
plt.barh(df2['Team_name'],df2['Unique_heros_played'],color='Black')

plt.subplot(5,1,4)
plt.title('Unique heros played')
plt.barh(df3['Team_name'],df3['Unique_heros_played'],color='Green')

plt.subplot(5,1,5)
plt.title('Unique heros played')
plt.barh(df3['Team_name'],df3['Unique_heros_played'],color='Purple')
print()
```

```python
plt.subplot(5,1,1)
plt.title('KDA')
plt.plot(df['Team_name'],df['KDA'],label=2016,color='Red')
plt.legend()
plt.subplot(5,1,2)
plt.title('KDA')
plt.plot(df1['Team_name'],df1['KDA'],label=2017,color='Blue')
plt.legend()
plt.subplot(5,1,3)
plt.title('KDA')
plt.plot(df2['Team_name'],df2['KDA'],label=2018,color='Black')
plt.legend()
plt.subplot(5,1,4)
plt.title('KDA')
plt.plot(df3['Team_name'],df3['KDA'],label=2019,color='Green')
plt.legend()
plt.subplot(5,1,5)
plt.title('KDA')
plt.plot(df4['Team_name'],df4['KDA'],label=2021,color= 'Purple')
plt.legend()
print()
```

```python
plt.subplot(5,1,1)
plt.title('Wins')
plt.bar(df['Team_name'],df['Wins'],label=2016,color='Red',width=0.4)
plt.legend()
plt.subplot(5,1,2)
plt.title('Wins')
plt.bar(df1['Team_name'],df1['Wins'],label=2017,color='Blue',width=0.4)
plt.legend()
plt.subplot(5,1,3)
plt.title('Wins')
plt.bar(df2['Team_name'],df2['Wins'],label=2018,color='Black',width=0.4)
plt.legend()
plt.subplot(5,1,4)
plt.title('Wins')
plt.bar(df3['Team_name'],df3['Wins'],label=2019,color='Green',width=0.4)
plt.legend()
plt.subplot(5,1,5)
plt.title('Wins')
plt.bar(df4['Team_name'],df4['Wins'],label=2021,color= 'Purple',width=0.4)
plt.legend()
print()
```
