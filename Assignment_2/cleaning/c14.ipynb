```python
import pandas as pd
```

```python
data_frame = pd.read_excel('DDW-0000C-14.xls')
```

```python
data_frame.drop(columns='Unnamed: 0',inplace=True)
data_frame.drop([0,1,2,3,4,5],inplace=True)
```

```python
data_frame.columns = ['State_code','District_code','Area_name','Age_grp','T_P','T_M','T_F','R_P','R_M','R_F','U_P','U_M','U_F']
```

```python
x = {'State_code' : 1, 'District_code' : 1, 'Area_name' : 1, 'Age_grp' : 1, 'T_P' : 1, 'T_M' : 1,
       'T_F' : 1, 'R_P' : 1, 'R_M' :1 , 'R_F' : 1, 'U_P' : 1, 'U_M' : 1, 'U_F' : 1 }
```

```python
merge = [(['30-34','35-39','40-44','45-49'],'30-49'),(['50-54','55-59','60-64','65-69'],'50-69'),(['70-74','75-79','80+'],'70+')]

col = ['T_P', 'T_M',
       'T_F', 'R_P', 'R_M', 'R_F', 'U_P', 'U_M', 'U_F']

state = list(data_frame['State_code'].unique())
```

```python

for s in state:
    temp  = data_frame[data_frame['State_code']==s]
    for i in merge:
        x = {'State_code' : 1, 'District_code' : 1, 'Area_name' : 1, 'Age_grp' : 1, 'T_P' : 1, 'T_M' : 1,
            'T_F' : 1, 'R_P' : 1, 'R_M' :1 , 'R_F' : 1, 'U_P' : 1, 'U_M' : 1, 'U_F' : 1 }
        for j in i:
            temp_age = temp[temp['Age_grp'].isin(i[0])]
            x['State_code'] = temp_age.iloc[0]['State_code']
            x['District_code'] = temp_age.iloc[0]['District_code']
            x['Area_name'] = temp_age.iloc[0]['Area_name']
            x['Age_grp'] = i[1]
            for k in col:
                x[k] = temp_age[k].sum()
        data_frame = data_frame.append(x,ignore_index=True)

        
```
